```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("https://raw.githubusercontent.com/victor-divo/project-komputer-grafis/main/UAS/stock.csv")
print(dataset.sort_values(by = 'purchase price', ascending=False).head())
```

```python
plt.clf()
plt.hist(dataset['purchase price'], bins=5, stacked=True, range=(1,800), color='red')
plt.title('Distribusi data Purchase Price', fontsize=16, color="black")
plt.xlabel("Purchase Price (Dollar)", fontsize = 14)
plt.ylabel("Jumlah Dari Stock Name", fontsize = 14)
plt.show()
```

```python
plt.clf()
dataset['current price formated'] = dataset['current price'].apply(lambda x: float(x.replace(',','.')))
plt.hist(dataset['current price formated'], bins=20, stacked=True, range=(1,1000), color='blue')
plt.title('Distribusi data Current Price', fontsize=16, color="black")
plt.xlabel("Current Price (Dollar)", fontsize = 14)
plt.ylabel("Jumlah Dari Stock Name", fontsize = 14)
plt.show()
```

```python
plt.clf()
dataset['gain/loss formated'] = dataset['gain/loss'].apply(lambda x: float(x.replace(',','.')))
plt.hist(dataset['gain/loss formated'], bins=20, stacked=True, range=(1,150000), color='green')
plt.title('Distribusi data Gain / Loss', fontsize=16, color="black")
plt.xlabel("Gain / Loss (Ribuan Dollar)", fontsize = 14)
plt.ylabel("Jumlah Dari Stock Name", fontsize = 14)
plt.xlim(xmin=0,xmax=150000)
labels, location = plt.xticks()
plt.xticks(labels, (labels/1000).astype(int))
plt.show()
```

```python
dataset['dividend/share'] = dataset['dividend/share'].apply(lambda x: x if isinstance(x, float) else float(x.replace(',','.')))
dataset_highest_dividen = dataset.sort_values(by = 'dividend/share', ascending=False).reset_index().head(5)
top_5_highest_dividen = dataset_highest_dividen[['stock name', 'symbol', 'dividend/share']]

print(top_5_highest_dividen)
```
