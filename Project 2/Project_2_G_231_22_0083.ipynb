```python
# Created by Victor Divo Mahendra G.231.22.0083
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv("https://raw.githubusercontent.com/victor-divo/project-komputer-grafis/main/Project%202/curah%20hujan.csv")
print(dataset.head())
```

```python
dataset['year'] = dataset['bulan'].apply(lambda x: int(datetime.datetime.strptime(x, "%Y-%m").strftime('%Y')))
dataset['year_group'] = pd.cut(dataset['year'], bins=[1981, 1991, 2001, 2011], labels=['1981-1990', '1991-2000', '2001-2010'], include_lowest=True, right=False)
dataset['month'] = dataset['bulan'].apply(lambda x: int(datetime.datetime.strptime(x, "%Y-%m").strftime('%m')))
print(dataset.head())
```

```python
plt.clf()
dataset.groupby(['month', 'year_group'])['curah hujan'].mean().unstack().plot(kind='bar')
plt.title("Data Curah hujan dikelompokkan berdasarkan bulan dan range tahun")
plt.grid(axis='y')
plt.xlabel("Bulan")
plt.ylabel("Curah Hujan (mm)")
plt.ylim(ymin=0, ymax=800)
plt.xticks(rotation=0)
plt.legend(ncols=3, frameon=False, loc='upper center')
```
