```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv")
dataset['order_month'] = dataset['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime('%Y-%m'))
dataset['gmv'] = dataset['item_price'] * dataset['quantity']
```

```python
plt.figure(figsize=(10, 5))
dataset[dataset['order_month']=='2019-12'].groupby(['order_date'])['customer_id'].nunique().plot(color='blue', marker='*', linewidth=3)
plt.title('Jumlah Pembeli Harian - Desember 2019', loc='center', pad=30, fontsize=20, color='blue')
plt.xlabel('Tanggal Pembelian', fontsize=16, color='black')
plt.ylabel('Unique Customers', fontsize=16, color='black')
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
```

```python
#Top 5 brands berdasarkan quantity
top_brands = (dataset[dataset['order_month']=='2019-12'].groupby('brand')['quantity'].sum().reset_index().sort_values(by='quantity',ascending=False).head(5))
# membuat list data baru, filter hanya yang bulan Desember 2019 dan top 5 brands
top_5_brands = dataset[(dataset['order_month']=='2019-12') & (dataset['brand'].isin(top_brands['brand'].to_list()))]
# tampilkan top brands
print(top_brands)
```

```python
top_5_brands.groupby(['order_date','brand'])['quantity'].sum().unstack().plot(marker='.', cmap='plasma')
plt.title('Kuantitas Penjualan Harian Desember 2019-Berdasarkan Brand',loc='center',pad=30, fontsize=15, color='blue')
plt.xlabel('Tanggal Pembelian', fontsize = 12)
plt.ylabel('Kuantitas',fontsize = 12)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)
plt.annotate('Lonjakan tertinggi', xy=(7, 310), xytext=(8, 300),
             weight='bold', color='blue',
             arrowprops=dict(arrowstyle='->',
                             connectionstyle="arc3",
                             color='red'))
plt.gcf().set_size_inches(10, 5)
plt.tight_layout()
```

```python
plt.clf()
top_5_brands.groupby('brand')['product_id'].nunique().sort_values(ascending=False).plot(kind='bar', color='red')
plt.title('Jumlah Produk terjual per brand (Desember 2019)',loc='center',pad=30, fontsize=15, color='blue')
plt.xlabel('Brand', fontsize = 15)
plt.ylabel('Jumlah Produk Terjual',fontsize = 15)
plt.ylim(ymin=0)
plt.xticks(rotation=0)
```

```python
top_5_brands_p_product = top_5_brands.groupby(['brand','product_id'])['quantity'].sum().reset_index()
top_5_brands_p_product['quantity_group'] = top_5_brands_p_product['quantity'].apply(lambda x: '>= 100' if x>=100 else '< 100')
top_5_brands_p_product.sort_values('quantity',ascending=False,inplace=True)
s_sort = top_5_brands_p_product.groupby('brand')['product_id'].nunique().sort_values(ascending=False)
top_5_brands_p_product.groupby(['brand','quantity_group'])['product_id'].nunique().reindex(index=s_sort.index, level='brand').unstack().plot(kind='bar', stacked=True)
plt.title('Jumlah Produk terjual per brand (Desember 2019)',loc='center',pad=30, fontsize=15, color='blue')
plt.xlabel('Brand', fontsize = 15)
plt.ylabel('Jumlah Produk Terjual',fontsize = 15)
plt.ylim(ymin=0)
plt.xticks(rotation=0)
```

```python
plt.figure(figsize=(10,5))
plt.hist(top_5_brands.groupby('product_id')['item_price'].median(), bins=10, stacked=True, range=(1,2000000), color='blue')
plt.title('Distribusi Median Harga per Produk\nTop 5 Brand (Des 2019)',fontsize=15, color='black')
plt.xlabel('Median Harga (Jutaan)', fontsize = 12)
plt.ylabel('Jumlah Produk Terjual',fontsize = 12)
plt.xlim(xmin=0,xmax=2000000)
```

```python
#agregat per product
data_per_product_top5brand_dec = top_5_brands.groupby('product_id').agg({'quantity': 'sum', 'gmv':'sum', 'item_price':'median'}).reset_index()
#scatter plot
plt.figure(figsize=(10,8))
plt.scatter(data_per_product_top5brand_dec['quantity'],data_per_product_top5brand_dec['gmv'], marker='*', color='green')
plt.title('Korelasi antara Quantity dan GMV per Produk\nTop 5 Brand (Desember 2019)',fontsize=15, color='black')
plt.xlabel('Quantity', fontsize = 12)
plt.ylabel('GMV (Jutaan)',fontsize = 12)
plt.xlim(xmin=0,xmax=300)
plt.ylim(ymin=0,ymax=200000000)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000).astype(int))
```

```python
plt.clf()
#agregat per product
data_per_product_top5brand_dec = top_5_brands.groupby('product_id').agg({'quantity': 'sum', 'gmv':'sum', 'item_price':'median'}).reset_index()
#scatter plot
plt.figure(figsize=(10,8))
plt.scatter(data_per_product_top5brand_dec['item_price'],data_per_product_top5brand_dec['quantity'], marker='o', color='black')
plt.title('Korelasi antara Quantity dan GMV per Produk\n Top 5 Brand (Desember 2019)',fontsize=15, color='blue')
plt.xlabel('Median Harga', fontsize = 12)
plt.ylabel('Quantity',fontsize = 12)
plt.xlim(xmin=0,xmax=2000000)
plt.ylim(ymin=0,ymax=250)
```
